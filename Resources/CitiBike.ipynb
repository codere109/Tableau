```python
import pandas as pd
```

```python
cb2019 = pd.read_csv('../Resources/june2019.csv')
cb2020 = pd.read_csv('../Resources/june2020.csv')
cb2021 = pd.read_csv('../Resources/june2021.csv')
cb2022 = pd.read_csv('../Resources/june2022.csv')
```

```python
#change gender column values to actual gender name M,F,Unknown for 2019 and 2020 dataset
cb2019.loc[cb2019['gender'] ==0,"gender"]="Unknown"
cb2019.loc[cb2019['gender'] ==1,"gender"]="Male"
cb2019.loc[cb2019['gender'] ==2,"gender"]="Female"

cb2020.loc[cb2020['gender'] ==0,"gender"]="Unknown"
cb2020.loc[cb2020['gender'] ==1,"gender"]="Male"
cb2020.loc[cb2020['gender'] ==2,"gender"]="Female"

```

```python
#rename column tripduration for 2019 & 2020
cb2019.rename(columns={"tripduration": "trip-duration(mins)"}, inplace=True)
cb2020.rename(columns={"tripduration": "trip-duration(mins)"}, inplace=True)
```
